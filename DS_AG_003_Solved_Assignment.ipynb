## Question 1: What is the difference between multithreading and multiprocessing?

### Answer

```python
'''Multithreading runs multiple threads within the same process and shares memory between them. It is useful for I/O-bound tasks such as file handling and network operations.

Multiprocessing runs multiple independent processes, each with its own memory space. It is useful for CPU-bound tasks because it can take advantage of multiple CPU cores.

Key Differences:
- Memory: Threads share memory; processes do not.
- Speed: Multiprocessing is generally better for CPU-intensive work.
- Communication: Threads communicate more easily than processes.'''
```

## Question 2: What are the challenges associated with memory management in Python?

### Answer

Some common memory management challenges in Python include:

1. Memory leaks caused by lingering object references.
2. Circular references that delay garbage collection.
3. High memory consumption with large datasets.
4. Fragmentation of memory.
5. Performance overhead from automatic garbage collection.

Python uses reference counting and a garbage collector to manage memory automatically.

## Question 3: Write a Python program that logs an error message to a log file when a division by zero exception occurs.

### Answer

```python
import logging

logging.basicConfig(
    filename='error.log',
    level=logging.ERROR,
    format='%(asctime)s - %(levelname)s - %(message)s'
)

try:
    result = 10 / 0
except ZeroDivisionError:
    logging.error("Division by zero occurred.")
    print("Error logged successfully.")
```

## Question 4: Write a Python program that reads from one file and writes its content to another file.

### Answer

```python
with open('source.txt', 'r') as source:
    content = source.read()

with open('destination.txt', 'w') as destination:
    destination.write(content)

print("File copied successfully.")
```

## Question 5: Write a program that handles both IndexError and KeyError using a try-except block.

### Answer

```python
my_list = [1, 2, 3]
my_dict = {"name": "Rahul"}

try:
    print(my_list[5])
    print(my_dict["age"])
except IndexError:
    print("IndexError: List index is out of range.")
except KeyError:
    print("KeyError: Key does not exist in dictionary.")
```

## Question 6: What are the differences between NumPy arrays and Python lists?

### Answer

```python
'''NumPy Arrays:
- Faster for numerical computations.
- Store elements of the same data type.
- Consume less memory.
- Support vectorized operations.

Python Lists:
- Can store different data types.
- More flexible but slower for large numerical operations.
- Require more memory.

NumPy is preferred for scientific computing and data analysis.
'''
```

## Question 10: Calculate the correlation matrix using Seaborn and visualize it with a heatmap.

### Answer

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('data.csv')

correlation_matrix = df.corr(numeric_only=True)

sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
